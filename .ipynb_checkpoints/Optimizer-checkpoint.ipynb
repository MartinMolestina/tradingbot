# Optimization


```python
# Relevant libraries
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn-v0_8")
import pandas_ta as ta
import warnings
warnings.filterwarnings("ignore")
from itertools import product
```

```python
# Function to backtest RSI strategy
def backtest(data, parameters, tc):
    data =data.copy()
    
    # prepare features
    data['RSI'] = data.ta.rsi(close='close', length=parameters[2])
    rsi_conditions = [
        (data['RSI'] > parameters[0]),
        (data['RSI'] <= parameters[0]) & (data['RSI'] > parameters[1]),
        (data['RSI'] <= parameters[1])]
    rsi_signals=[-1,0,1]
    data["rsi_indicator"] = np.select(rsi_conditions,rsi_signals)

    # Calculate the hystorical returns
    data["log_return"] = ta.log_return(close=data["close"])

    # Define what asset is the strategy holding (BTC or USDT)
    data['asset'] = calculate_asset_holding(data['rsi_indicator'].tolist())

    # Backtest
    data['trading'] = np.where(data['asset'] == 'BTC', 1, 0)
    data['strategy'] = data['trading'].shift(1) * data['log_return']
    data["cstrategy"] = data["strategy"].cumsum().apply(np.exp) 
    data["creturns"] = data['log_return'].cumsum().apply(np.exp)

    # Return strategy multiple after backtest training cycle
    return data["cstrategy"][data["cstrategy"].size-1]

# Function to calculate asset holding
def calculate_asset_holding(signals):
    asset = 'USDT'
    asset_list = [asset]

    for signal in signals:
        if signal == 1:
            asset = 'BTC'
        elif signal == -1:
            asset = 'USDT'
        asset_list.append(asset)

    return asset_list[:-1]  # Return the last element (only for alignment)
```

```python
# Define initial paramters
RSI_sell_limit = 75
RSI_buy_limit = 30
RSI_length = 10

parameters = [RSI_sell_limit, RSI_buy_limit, RSI_length]

tc = -0.00085

data = pd.read_csv('bitcoin.csv')  
```

```python
backtest(data = data, parameters = parameters, tc = tc)
```

```python
# Define initial paramters ranges
RSI_sell_limit = range(70,90,1)
RSI_buy_limit = range(15,40,1)
RSI_length = range(5,20,1)

combinations = list(product(RSI_sell_limit, RSI_buy_limit, RSI_length))

tc = -0.00085

len(combinations)
combinations[0]
```

```python
results = []
for comb in combinations:
    results.append(backtest(data = data, parameters = comb, tc = tc))
```

```python
many_results =  pd.DataFrame(data = combinations, columns = ["RSI sell", "RSI buy", "RSI length"])
many_results["performance"] = results
```
